```python
pnl = pd.read_csv("able.csv", sep = ";")
pnl.head()
```

```python
# List of products
products = ['CHOCOLATE',  'GIFT_BASKET', 'ORCHIDS', 'STARFRUIT', 'AMETHYSTS']

# Create a figure with 3 subplots
fig, axes = plt.subplots(len(products), 1, figsize=(10, 15))

# Loop through each product and plot the cumulative P&L
for i, product in enumerate(products):
    product_data = pnl[pnl['product'] == product]
    axes[i].plot(product_data['timestamp'], product_data['profit_and_loss'])
    axes[i].set_title(f'Cumulative P&L for {product}')
    axes[i].set_xlabel('Timestamp')
    axes[i].set_ylabel('Cumulative P&L')

plt.tight_layout()
plt.show()
```
